# Defanging an IP Address

```python
def defangIPaddr(address):
    res = ""
    for i in address:
        if i == ".":
            res += "[.]"
        else:
            res += i
    return res
```

```python
address = "1.1.1.1"
defangIPaddr(address)
```

```python
address = "255.100.50.0"
defangIPaddr(address)
```

# Find Numbers with Even Number of Digits

```python
def findNumbers(nums):
    c=0
    for i in nums:
        temp=i
        digits=0
        while temp>0:
            digits+=1
            temp=temp//10
        if digits%2==0:
            c+=1
    return c
```

```python
nums = [12,345,2,6,7896]
findNumbers(nums)
```

```python
nums = [555,901,482,1771]
findNumbers(nums)
```

# Number of Good Pairs

```python
def numIdenticalPairs(nums):
    c = 0
    for i in range(len(nums)):
        for j in range(i + 1, len(nums)):
            if nums[i] == nums[j]:
                c += 1

    return c
```

```python
nums = [1,2,3,1,1,3]
numIdenticalPairs(nums)
```

```python
nums = [1,1,1,1]
numIdenticalPairs(nums)
```

```python
nums = [1,2,3]
numIdenticalPairs(nums)
```

# How Many Numbers Are Smaller Than the Current Number

```python
def smallerNumbersThanCurrent(nums):
    out=[]
    for i in range (len(nums)):
        c=0
        for j in range (len(nums)):
            if j!=i and nums[j]< nums[i]:
                c+=1
        out.append(c)
    return out
```

```python
nums = [8,1,2,2,3]
smallerNumbersThanCurrent(nums)
```

```python
nums = [6,5,4,8]
```

```python
smallerNumbersThanCurrent(nums)
```

```python
nums = [7,7,7,7]
smallerNumbersThanCurrent(nums)
```

# Subtract the Product and Sum of Digits of an Integer

```python
def subtractProductAndSum(n):
    pod = 1
    sod = 0
    temp = n

    while temp > 0:
        digit = temp % 10
        pod *= digit
        sod += digit
        temp = temp // 10

    return pod - sod
```

```python
n = 234
subtractProductAndSum(n)
```

```python
n = 4421
subtractProductAndSum(n)
```

# XOR Operation in an Array

```python
def xorOperation(n, start):
    res = 0
    nums = []

    for i in range(n):
        val = start + 2 * i
        nums.append(val)

    for j in nums:
        res = res ^ j

    return res
```

```python
n = 5
start = 0
xorOperation(n,start)
```

```python
n = 4
start = 3
xorOperation(n,start)
```
