# Running sum of 1D array

```python
def runningSum(nums):
    out=[]
    total=0
    for i in nums:
        total+=i
        out.append(total)
    return out
```

```python
nums = [3,1,2,10,1]
runningSum(nums)
```

# Shuffle the array

```python
def shuffle(nums, n):
    res = []
    for i in range(n):
        res.append(nums[i])  
        res.append(nums[i + n])
    
    return res
```

```python
nums = [1,1,2,2]
n = 2
shuffle(nums,n)
```

# Kids With the Greatest Number of Candies

```python
def kidsWithCandies(candies, extraCandies):
    res = []
    maximum_candies = max(candies)
    for i in candies:
        if i + extraCandies >= maximum_candies:
            res.append(True)
        else:
            res.append(False)
    
    return res
```

```python
candies = [4,2,1,1,2]
extraCandies = 1
kidsWithCandies(candies,extraCandies)
```
